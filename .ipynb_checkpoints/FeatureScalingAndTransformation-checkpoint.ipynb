```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Floating points being rounded of to three places after the decimal
np.set_printoptions(precision=3)
```

```python
diabetes = pd.read_csv('datasets/diabetes_processed_1.csv')
diabetes.head()
```

```python
features = diabetes.drop('Outcome',axis=1)
target = diabetes['Outcome']
features.describe()
```

# Feature Transformation Techiniques

```python
#One way to transform data is the MinMaxScaler [Scaling]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_features = scaler.fit_transform(features)
scaled_features = pd.DataFrame(scaled_features, columns = features.columns)
scaled_features.describe()
```

```python
# Another transformatin technique used on data StandardScaler [Standardization]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit(features)
standardized_features = scaler.transform(features)
standardized_features = pd.DataFrame(standardized_features, columns=features.columns)
standardized_features.describe()
```

```python
standardized_features.boxplot(figsize=(10,8), rot=45)
plt.show()
```

```python
# Normalization
from sklearn.preprocessing import Normalizer
# norm=l1 [Sum of the abs values of features will be equal to 1]
normalizer = Normalizer(norm='l1')
normalized_features=normalizer.fit_transform(features)
l1_normalized_features = pd.DataFrame(normalized_features, columns=features.columns)
# l1_normalized_features.describe()
l1_normalized_features.iloc[0]
l1_normalized_features.iloc[0].abs().sum()
```
